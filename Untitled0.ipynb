```python
import pandas as pd
import json
import torch
from datasets import load_dataset
from transformers import BartForConditionalGeneration, BartTokenizer, Seq2SeqTrainer, Seq2SeqTrainingArguments
from google.colab import files

# Step 1: Upload the file
uploaded = files.upload()

# Step 2: Extract the file name from the uploaded dictionary
file_path = list(uploaded.keys())[0]

# Step 3: Load the Excel file
data = pd.read_excel(file_path)

# Step 4: Prepare input and output columns
data["input"] = data["abstract"]  # Adjust column names if needed
data["output"] = data["title"]    # Adjust column names if needed
processed_data = data[["input", "output"]]

# Step 5: Save the processed data to a JSONL file
output_file = "fine_tuning_data.jsonl"
with open(output_file, "w") as f:
    for _, row in processed_data.iterrows():
        json.dump({"input": row["input"], "output": row["output"]}, f)
        f.write("\n")

# Step 6: Load dataset from JSONL file
dataset_path = 'fine_tuning_data.jsonl'
dataset = load_dataset("json", data_files={"train": dataset_path})

# Step 7: Select the model and tokenizer
model_name = 'facebook/bart-base'
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)

# Step 8: Tokenize the dataset
def preprocess_function(examples):
    inputs = [ex for ex in examples["input"]]
    targets = [ex for ex in examples["output"]]
    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding="max_length")
    labels = tokenizer(targets, max_length=128, truncation=True, padding="max_length")
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Preprocess the dataset
train_dataset = dataset["train"].map(preprocess_function, batched=True)

# Step 9: Set device to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Step 10: Define training arguments
training_args = Seq2SeqTrainingArguments(
    output_dir="./bart_results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    weight_decay=0.01,
    logging_dir="./bart_logs",
    save_steps=500,
    save_total_limit=2,
    num_train_epochs=3,
    predict_with_generate=True
)

# Step 11: Initialize the Seq2SeqTrainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=train_dataset,
    tokenizer=tokenizer
)

# Step 12: Train the model
trainer.train()

# Step 13: Save the fine-tuned model and tokenizer
model.save_pretrained("./bart_fine_tuned_model")
tokenizer.save_pretrained("./bart_fine_tuned_model")

# Step 14: Optional: Test the model after fine-tuning
input_text = "Example abstract to generate title for."
inputs = tokenizer(input_text, return_tensors="pt").to(device)
generated_ids = model.generate(inputs['input_ids'], max_length=50, num_beams=5, early_stopping=True)
generated_text = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
print(f"Generated Title: {generated_text}")
```

```python
from rouge_score import rouge_scorer

# Step 15: Evaluate the model using ROUGE scores

# Load test data (using the same dataset here, but you can use a separate test set)
input_abstracts = dataset["train"]["input"]  # You can replace with a separate test dataset if needed
actual_titles = dataset["train"]["output"]
# Generate titles using the fine-tuned BART model for a subset of inputs
generated_titles = []
num_samples_to_generate = 5  # Limit to 5 inputs (adjust as needed)

# Loop through a limited number of samples
for i, abstract in enumerate(input_abstracts[:num_samples_to_generate]):
    # Tokenize the abstract
    inputs = tokenizer(abstract, return_tensors="pt", max_length=512, truncation=True, padding="longest")

    # Generate title using the BART model
    with torch.no_grad():
        generated_ids = model.generate(inputs["input_ids"].to(device), max_length=64, num_beams=4, early_stopping=True)

    # Decode the generated token IDs
    generated_title = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
    generated_titles.append(generated_title)

# Display the generated titles for the limited number of inputs
for idx, (abstract, generated_title) in enumerate(zip(input_abstracts[:num_samples_to_generate], generated_titles)):
    print(f"Sample {idx + 1}:")
    print(f"Input (Abstract): {abstract}")
    print(f"Generated Title: {generated_title}")
    print("-" * 80)



# Step 16: Initialize ROUGE scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

# Calculate ROUGE scores for all samples
rouge_scores = []
for gen, actual in zip(generated_titles, actual_titles):
    score = scorer.score(actual, gen)
    rouge_scores.append(score)

# Step 17: Calculate average ROUGE scores
avg_rouge_1 = sum([score['rouge1'].fmeasure for score in rouge_scores]) / len(rouge_scores)
avg_rouge_2 = sum([score['rouge2'].fmeasure for score in rouge_scores]) / len(rouge_scores)
avg_rouge_L = sum([score['rougeL'].fmeasure for score in rouge_scores]) / len(rouge_scores)

print(f"Average ROUGE-1: {avg_rouge_1}")
print(f"Average ROUGE-2: {avg_rouge_2}")
print(f"Average ROUGE-L: {avg_rouge_L}")

# Step 18: Display evaluation results for each sample (input, generated title, actual title, ROUGE scores)
for idx, (abstract, generated, actual, score) in enumerate(zip(input_abstracts, generated_titles, actual_titles, rouge_scores)):
    print(f"Sample {idx + 1}:")
    print(f"Input (Abstract): {abstract}")
    print(f"Generated Title: {generated}")
    print(f"Actual Title: {actual}")
    print(f"ROUGE-1: {score['rouge1'].fmeasure}")
    print(f"ROUGE-2: {score['rouge2'].fmeasure}")
    print(f"ROUGE-L: {score['rougeL'].fmeasure}")
    print("-" * 80)
```
