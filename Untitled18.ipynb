```python
import pandas as pd
import json
import torch
from datasets import load_dataset
from transformers import BartForConditionalGeneration, BartTokenizer, Seq2SeqTrainer, Seq2SeqTrainingArguments
from google.colab import files

# Step 1: Upload the file
uploaded = files.upload()

# Step 2: Extract the file name from the uploaded dictionary
file_path = list(uploaded.keys())[0]

# Step 3: Load the Excel file
data = pd.read_excel(file_path)

# Step 4: Prepare input and output columns
data["input"] = data["abstract"]  # Adjust column names if needed
data["output"] = data["title"]    # Adjust column names if needed
processed_data = data[["input", "output"]]

# Step 5: Save the processed data to a JSONL file
output_file = "fine_tuning_data.jsonl"
with open(output_file, "w") as f:
    for _, row in processed_data.iterrows():
        json.dump({"input": row["input"], "output": row["output"]}, f)
        f.write("\n")

# Step 6: Load dataset from JSONL file
dataset_path = 'fine_tuning_data.jsonl'
dataset = load_dataset("json", data_files={"train": dataset_path})

# Step 7: Select the model and tokenizer
model_name = 'facebook/bart-base'
tokenizer = BartTokenizer.from_pretrained(model_name)
model = BartForConditionalGeneration.from_pretrained(model_name)

# Step 8: Tokenize the dataset
def preprocess_function(examples):
    inputs = [ex for ex in examples["input"]]
    targets = [ex for ex in examples["output"]]
    model_inputs = tokenizer(inputs, max_length=512, truncation=True, padding="max_length")
    labels = tokenizer(targets, max_length=128, truncation=True, padding="max_length")
    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

# Preprocess the dataset
train_dataset = dataset["train"].map(preprocess_function, batched=True)

# Step 9: Set device to GPU if available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Step 10: Define training arguments
training_args = Seq2SeqTrainingArguments(
    output_dir="./bart_results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    weight_decay=0.01,
    logging_dir="./bart_logs",
    save_steps=500,
    save_total_limit=2,
    num_train_epochs=3,
    predict_with_generate=True
)

# Step 11: Initialize the Seq2SeqTrainer
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=train_dataset,
    tokenizer=tokenizer
)

# Step 12: Train the model
trainer.train()

# Step 13: Save the fine-tuned model and tokenizer
model.save_pretrained("./bart_fine_tuned_model")
tokenizer.save_pretrained("./bart_fine_tuned_model")

# Step 14: Optional: Test the model after fine-tuning
input_text = "Example abstract to generate title for."
inputs = tokenizer(input_text, return_tensors="pt").to(device)
generated_ids = model.generate(inputs['input_ids'], max_length=50, num_beams=5, early_stopping=True)
generated_text = tokenizer.decode(generated_ids[0], skip_special_tokens=True)
print(f"Generated Title: {generated_text}")
```

```python
import torch
from transformers import BartForConditionalGeneration, BartTokenizer
from bart_score import BARTScorer

# Step 1: Load the fine-tuned BART model and tokenizer
model_path = "./bart_fine_tuned_model"  # Path to your fine-tuned model
tokenizer = BartTokenizer.from_pretrained(model_path)
model = BartForConditionalGeneration.from_pretrained(model_path)

# Step 2: Initialize BARTScorer
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
scorer = BARTScorer(device=device, checkpoint="facebook/bart-large-cnn")

# Step 3: Set device for the BART model
model.to(device)

# Step 4: Prepare test data (abstracts and expected titles)
test_data = [
    {"abstract": "The study aims to explore advancements in artificial intelligence.", "expected_title": "Advancements in AI"},
    {"abstract": "The research focuses on natural language processing techniques.", "expected_title": "NLP Techniques in Research"},
    # Add more test samples as needed
]

# Step 5: Generate titles
generated_titles = []
expected_titles = []

for example in test_data:
    abstract = example["abstract"]
    expected_title = example["expected_title"]

    # Tokenize the input (abstract)
    inputs = tokenizer(
        abstract, return_tensors="pt", truncation=True, padding="max_length", max_length=512
    ).to(device)

    # Generate title using the fine-tuned BART model
    generated_ids = model.generate(inputs['input_ids'], max_length=50, num_beams=5, early_stopping=True)
    generated_title = tokenizer.decode(generated_ids[0], skip_special_tokens=True)

    # Append generated and expected titles
    generated_titles.append(generated_title)
    expected_titles.append(expected_title)

# Step 6: Evaluate using BARTScore
bart_scores = scorer.score(generated_titles, expected_titles)

# Step 7: Print results
for i, (gen, ref, score) in enumerate(zip(generated_titles, expected_titles, bart_scores)):
    print(f"Example {i+1}:")
    print(f"Generated Title: {gen}")
    print(f"Expected Title: {ref}")
    print(f"BARTScore: {score:.4f}\n")

# Step 8: Calculate the average BARTScore
average_score = sum(bart_scores) / len(bart_scores)
print(f"Average BARTScore: {average_score:.4f}")
```
