Write a python program to collect text data from **either of the following sources** and save the data into a **csv file:**

(2) Collect the top 1000 User Reviews of a movie recently in 2023 or 2024 (you can choose any movie) from IMDB. [If one movie doesn't have sufficient reviews, collect reviews of atleast 2 or 3 movies]

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import time
```

```python
import requests
from bs4 import BeautifulSoup
url = "https://www.imdb.com/title/tt10698680/reviews/?ref_=tt_ov_ql_2"
headers = {
    "User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:123.0) Gecko/20100101 Firefox/123.0",
    "Accept": "text/html",
    "Connection": "keep-alive"
}
resp1 = requests.get(url, headers=headers)
res_resu1 = BeautifulSoup(resp1.content, 'html.parser')
result1 = res_resu1.find_all(class_='review-container')
final_result=result1

```

```python
all_list = []
def col_name(item):
  username = item.a.get_text()
  return username
def rev_tex(item):
  rev_tex = item.find(class_='text').get_text()
  return rev_tex
def rat_ing(item):
  rating_element = item.find(class_='rating-other-user-rating')
  if rating_element:
        rating = rating_element.get_text().strip()
        return rating
  else:
        return "N/A"
for item in final_result:
    user_name=col_name(item.find(class_='display-name-link'))
    review=rev_tex(item)
    rating=rat_ing(item)
    # joining all the three user,review,rating
    all_list.append({
        'username': user_name,
        'review': review,
        'rating': rating
    })
for entry in all_list:
    print(f"rating: {entry['rating']}")
    print(f"username: {entry['username']}")
    print(f"review: {entry['review']}")
    print()
import csv
with open ('gnaneshwar.csv', mode='w', newline='') as file:
   writer = csv.DictWriter(file, fieldnames=['username', 'review', 'rating'])
   writer.writeheader ()
   writer.writerows (all_list)
   print("CSV has created succesfully and downloaded")
```

# Question 2 (30 points)

Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the stopwords list.

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re
import pandas as pd
nltk.download ('punkt')
nltk.download ('stopwords')
nltk.download( 'wordnet')
stemmer= PorterStemmer()
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words( 'english' ))
```

```python
import re

def clean_t(text):
    revi_tex= re.sub(r'[^\w\s]', '', text)
    revi_tex = re.sub(r'\d+', '', revi_tex)
    revi_tex = revi_tex.lower()
    fil_words = [word for word in revi_tex.split() if word.lower() not in stop_words]
    return ' '.join(fil_words)
cleaned_text = []
for item in all_list:
    cleaned_text.append(clean_t(item['review']))
print(all_list[0]['review'])
print(cleaned_text[0])
for i, item in enumerate(all_list):
     item[ 'cleaned_review'] = cleaned_text[1]
with open ('gnaneshwar_filter.csv', mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=['rating', 'review', 'username', 'cleaned_review' ])
    writer.writeheader()
    writer.writerows(all_list)
    print("CSV file created.")
```

# Question 3 (30 points)

Write a python program to **conduct syntax and structure analysis of the clean text** you just saved above. The syntax and structure analysis includes:

(1) **Parts of Speech (POS) Tagging:** Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

(2) **Constituency Parsing and Dependency Parsing:** print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) **Named Entity Recognition:** Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
import spacy
from spacy import displacy
nlp = spacy. load ("en_core_web_sm")
cleaned_text1 = pd.read_csv('gnaneshwar_filter.csv')
cleaned_review = cleaned_text1['cleaned_review']
from collections import Counter
pos_counts = Counter({'Nouns': 0, 'Verbs': 0, 'Adjectives': 0, 'Adverbs': 0})
#1.pos tag
def pos_tag(item):
    doc = nlp(item)
    for tokens in doc:
        match tokens.pos_:
            case 'NOUN':
                pos_counts['Nouns'] += 1
            case 'VERB':
                pos_counts['Verbs'] += 1
            case 'ADJ':
                pos_counts['Adjectives'] += 1
            case 'ADV':
                pos_counts['Adverbs'] += 1
counts_review = cleaned_review.apply(pos_tag)
print("\nPOS Counts:")
print(f"Total Nouns: {pos_counts['Nouns']}")
print(f"Total Verbs: {pos_counts['Verbs']}")
print(f"Total Adjectives: {pos_counts['Adjectives']}")
print(f"Total Adverbs: {pos_counts['Adverbs']}")
#2.parsing
def deped_pars(item):
    print(item)
    dokumt = nlp(item)
    displacy.render(dokumt, style='dep', jupyter=True)

deped_pars(cleaned_review[0])

def ner_ana(texts):
    tags = Counter()
    for text in texts:
        dokumt = nlp(text)
        for ent in dokumt.ents:
            tags[ent.label_] += 1
    for ent in tags:
        print(f"{ent}: {tags[ent]}")
ner_results = ner_ana(cleaned_review)

#printing the entities below
#print(f'the entities found are {ner_results}')
```
